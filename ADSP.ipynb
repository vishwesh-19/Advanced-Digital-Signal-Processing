```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import signal

f = 5
fs = 100
n = np.linspace(-20,20,41)
x = np.sin(2*np.pi*f*n/fs)
plt.stem(x)
plt.xlabel('n')
plt.ylabel('x[n]', rotation=0)
```

```python
X = scipy.fft(x)
X = abs(X)
plt.plot(X)
```

```python
# Upsampling
M=2
up = np.zeros((M*len(n),), dtype=float)
for i in range(0,len(up), 2):
  up[i] = n[i//2]

#plt.stem(up)
x_up = np.sin(2*np.pi*f*up/fs)
plt.stem(up, x_up)
plt.xlabel('n')
plt.ylabel('upsample(n)')
```

```python
Xup = scipy.fft(x_up)
plt.plot(abs(Xup))
```

```python
# Downsampling
down = np.zeros((int(len(n)/M)+1,), dtype=float)
for j in range(0, len(down)):
  down[j] = n[2*j]

x_down = np.sin(2*np.pi*f*down/fs)
plt.stem(down, x_down)
plt.xlabel('n')
plt.ylabel('downsample(n)')
#print(len(down))
```

```python
Xdown = scipy.fft(x_down)
plt.plot(abs(Xdown))
```
